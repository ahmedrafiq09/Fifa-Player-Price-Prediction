```python
# Import All The Important Libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Load The Dataset And Save it Into Variable
df = pd.read_csv("player_stats.csv", encoding='ISO-8859-1')
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Iterate over columns and apply label encoding only to non-numeric columns
for column in df.columns:
    if df[column].dtype == 'object':  # Check if the column is non-numeric
        try:
            # Convert the column to strings before label encoding
            df[column] = le.fit_transform(df[column].astype(str))
        except Exception as e:
            print(f"Error encoding column '{column}': {e}")
```

```python
# Data Preprocessing: Remove null values and label encode
df.fillna(0, inplace=True)
le = LabelEncoder()
```

```python
# Label encode columns with 'object' data type
object_columns = df.select_dtypes(include=['object']).columns
for column in object_columns:
    df[column] = le.fit_transform(df[column].astype(str))
```
