```python
# Import All The Important Libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Load The Dataset And Save it Into Variable
df = pd.read_csv("player_stats.csv", encoding='ISO-8859-1')
df
```

```python
# Check for null values in the entire dataset
null_values = df.isnull().sum()

# Display the columns with null values and their counts
columns_with_null = null_values[null_values > 0]
if not columns_with_null.empty:
    print("Columns with null values:")
    print(columns_with_null)
else:
    print("No null values in the dataset.")
```

```python
# Specify the column name with null values
column_to_fill = 'marking'

# Fill null values in the specified column with a value (e.g., 0)
df[column_to_fill].fillna(0, inplace=True)

# Check if there are still any null values in the specified column
null_values_after_fill = df[column_to_fill].isnull().sum()

if null_values_after_fill == 0:
    print(f"Null values in '{column_to_fill}' column filled successfully.")
else:
    print(f"Failed to fill null values in '{column_to_fill}' column.")
```

```python
# Check for null values in the entire dataset
null_values = df.isnull().sum()

# Display the columns with null values and their counts
columns_with_null = null_values[null_values > 0]
if not columns_with_null.empty:
    print("Columns with null values:")
    print(columns_with_null)
else:
    print("No null values in the dataset.")
```

```python
# Remove the currency symbol ($) and dots from the 'value' column
df['value'] = df['value'].replace('[\$,]', '', regex=True).replace('\.', '', regex=True).astype(float) / 1e6
df
```
